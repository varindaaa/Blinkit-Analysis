### **BUSINESS REQUIREMENTS**

#### **KPI's REQUIREMENTS**

```python
  #Total Sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df['Sales'].mean()

#No of Items Sold
no_of_items_sold = df['Sales'].count()

#Average Ratings
avg_ratings = df['Rating'].mean()

#Display

print(f"Total Sales : ${total_sales:,.0f}")
print(f"Average Sales : ${avg_sales:,.0f}")
print(f"No of Items Sold : {no_of_items_sold:,.0f}")
print(f"Average Ratings : {avg_ratings:,.1f}")
```

#### **CHARTS REQUIREMENTS**

##### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

n =2 
colors = ['#cda900','#568949']

plt.pie(sales_by_fat , labels = sales_by_fat.index,
                      autopct = '%.1f%%' ,
                       startangle = 90,
                       colors = colors)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

##### **Total Sales by Item Type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)

colors = sns.color_palette("crest", n_colors=len(sales_by_type))

plt.figure(figsize = (10,6))
bars = plt.bar(sales_by_type.index, sales_by_type.values, color = colors)

plt.xticks(rotation = 90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
    f'{bar.get_height():,.0f}' , ha = 'center' , va = 'bottom' , fontsize = 8)


plt.tight_layout()
plt.show()
```

##### **Fat Content by Outlet for Total Sales**

```python
grouped = df.groupby(['Outlet Location Type' , 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular' ,'Low Fat']]

colors = sns.color_palette("Paired")

ax = grouped.plot(kind = 'bar' , figsize = (8,5),color = colors , title = 'Outlet Tier by Item Fat Content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sale')
plt.legend(title = 'Item Fat Content')
plt.tight_layout()
plt.show()
```

##### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index , sales_by_year.values , color = 'pink',marker = "o" ,markerfacecolor = 'deeppink',markeredgecolor = 'deeppink', linestyle = '-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

for x,y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}',ha = 'center' , va = 'bottom' , fontsize =8)


plt.tight_layout()
plt.show()
```

##### **Sales by Outlet Size**

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

colors = sns.color_palette("rocket_r")

plt.figure(figsize = (4,4))
plt.pie(sales_by_size, labels = sales_by_size.index ,colors = colors, autopct = '%1.1f%%', startangle = 90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

##### **Sales by Outlet Location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

num_bars = sales_by_location['Outlet Location Type'].nunique()

plt.figure(figsize =(8,3)) 
ax = sns.barplot(x='Sales' , 
                 y = 'Outlet Location Type', 
                 hue = "Outlet Location Type",
                 palette =sns.color_palette("cubehelix" , n_colors = num_bars), 
                 data = sales_by_location)

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location TYpe')

plt.tight_layout()
plt.show()
```

```python
pivot_data = df.pivot_table(index='Item Type', columns='Item Fat Content', values='Rating', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, cmap='flare', fmt=".1f" )
plt.title('Average Rating by Item Type and Fat Content')
plt.xlabel('Item Fat Content')
plt.ylabel('Item Type')
plt.show()
```

## Conclusion: Blinkit Sales Analysis

In this project, we analyzed Blinkit's retail dataset to uncover insights into product performance, outlet characteristics, and customer preferences. The dataset included various dimensions such as product type, fat content, outlet attributes, and customer ratings.

**Key findings:**

- **Tier3 outlet locations** generate the highest sales, as visualized in the bar chart.
- **Low Fat items** generally receive slightly better **customer ratings**, indicating a health-conscious preference.
- Categories like **Dairy**, **Frozen Foods**, and **Baking Goods** are among the top-rated item types.
- **Larger outlets** and **Supermarket Type 1** formats drive higher sales, pointing to scale and store type as major factors.

**Recommendations:**

- Focus marketing and stock planning on **urban locations** and **top-performing product categories**.
- Expand **Low Fat product lines** to increase customer satisfaction.
- Consider opening or expanding **supermarket-style outlets** in high-traffic regions.

**Limitations:**

- The analysis is based on static data and does not include time-based trends or seasonal patterns.

**Future Scope:**

- Add **time-series forecasting** for sales.
- Perform **market basket analysis** to discover product bundling opportunities.
- Use clustering or segmentation to better understand **outlet-level performance**.
